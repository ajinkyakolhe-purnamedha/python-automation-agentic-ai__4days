## Section 1 — Variables

Build up the toolkit: **variables & operations** → **types** → **type hints** (just a hint, not enforced) → **list & dict** → **using** them (index/key + methods) → and finally combine every type into one **account dict**, the domain object we carry through the whole course.

```python
# 1.1 · Variables & operations — a name points at a value; Python infers the type, then you operate.
owner = "Ada"
balance = 1500.0
rate = 0.04
balance = balance + balance * rate    # arithmetic on numbers
label = "owner: " + owner             # + joins strings
wealthy = balance > 5000              # comparison -> a bool
print(label, "| balance:", round(balance, 2), "| wealthy?", wealthy)
```

```python
# 1.2 · Types — type() reveals the kind Python inferred. These are Python's core types.
print(type("Ada"))     # <class 'str'>
print(type(1))         # <class 'int'>
print(type(1500.0))    # <class 'float'>
print(type(True))      # <class 'bool'>
print(type(None))      # <class 'NoneType'>
```

```python
# 1.3 · Type hints are JUST hints — a variable can hold any type and be rebound at runtime.
x = 1500.0          # float now...
x = "frozen"        # ...str later — Python allows it
print(x)

balance: float = 1500.0   # the hint documents intent; it won't stop balance = "frozen"
print(balance)
```

```python
# 1.4 · Advanced variables: a list holds many in order; a dict holds named fields by key.
tags: list[str] = ["primary", "online"]
acct_min: dict = {"owner": "Ada", "balance": 1500.0}
print(tags)
print(acct_min)
```

```python
# 1.5 · Using them: index a list, key a dict, and call their methods.
print(tags[0], "| len:", len(tags))                          # primary | 2
print(acct_min["owner"], "| get:", acct_min.get("missing"))  # Ada | None (safe, no KeyError)
print(list(acct_min.keys()), list(acct_min.values()))        # field names ; values
for key, value in acct_min.items():                          # items() -> (key, value) pairs
    print("  ", key, "=", value)
```

```python
# 1.6 · Combine every type into ONE dict — a bank account. Carried through the whole course.
acct = {
    "id": 1,                       # int
    "owner": "Ada",                # str
    "account_type": "savings",     # str
    "balance": 1500.0,             # float
    "is_active": True,             # bool
    "tags": ["primary", "online"], # list[str]
}
print(acct)
```

## Section 2 — Control flow

Now we make decisions and repeat work: **truthiness** (empty/zero values are falsy) → **branching** with `if/elif/else` → **loops** (`for` over a list, a dict's `.items()`, a counter via `enumerate`, a `range`; `while` until a condition flips).

```python
# 2.1 · Truthiness — empty/zero values are falsy; test a value straight in `if`.
print(bool(0), bool(""), bool([]), bool(None))   # all False
if not acct["tags"]:        # acct from Section 1; an empty list would be falsy
    print("no tags yet")
else:
    print("tags:", acct["tags"])
```

```python
# 2.2 · Branching — if/elif/else picks exactly one branch.
accounts = [
    {"id": 1, "owner": "Ada", "balance": 1500.0, "is_active": True},
    {"id": 2, "owner": "Lin", "balance": 0.0,    "is_active": False},
    {"id": 3, "owner": "Sam", "balance": 40.0,   "is_active": True},
]
a = accounts[2]
if not a["is_active"]:    label = "inactive"
elif a["balance"] < 100:  label = "low balance"
else:                     label = "healthy"
print(a["owner"], "->", label)   # Sam -> low balance
```

```python
# 2.3 · Loops — for over a list, a dict's items, a counter (enumerate), a range; while until false.
for acc in accounts:                       # over a list
    print("  list:", acc["owner"])
for key, value in accounts[0].items():     # over a dict's (key, value) pairs
    print("  item:", key, "=", value)
for i, acc in enumerate(accounts):         # i is the counter: 0, 1, 2
    print("  #", i, acc["owner"])
for n in range(3):                         # 0, 1, 2
    print("  range:", n)
sam = accounts[2]
while sam["balance"] < 100:                # repeat until it clears 100
    sam["balance"] += 25
print("Sam after top-ups:", sam["balance"])  # 115.0
```

## Section 3 — Functions

Wrap logic so it's reusable: **arguments, default values, return** → **type hints on functions** (document the signature; not enforced) → **flexible args** (`*args`, `**kwargs`).

```python
# 3.1 · Arguments, a default value, and a return.
accounts = [
    {"id": 1, "owner": "Ada", "balance": 1500.0, "is_active": True},
    {"id": 2, "owner": "Lin", "balance": 200.0,  "is_active": False},
    {"id": 3, "owner": "Sam", "balance": 40.0,   "is_active": True},
]
def total_balance(accounts, only_active=True):     # only_active defaults to True
    if only_active:
        accounts = [a for a in accounts if a["is_active"]]
    return sum(a["balance"] for a in accounts)
print(total_balance(accounts))                     # 1540.0 (positional -> default True)
print(total_balance(accounts, only_active=False))  # 1740.0 (keyword arg)
```

```python
# 3.2 · Type hints on functions — document the signature. NOT enforced: a wrong type still runs.
def total_balance_typed(accounts: list[dict], only_active: bool = True) -> float:
    rows = [a for a in accounts if a["is_active"]] if only_active else accounts
    return sum(a["balance"] for a in rows)
print(total_balance_typed(accounts))               # 1540.0 — hints didn't change behaviour
```

```python
# 3.3 · Flexible args: *args (extra positionals -> tuple), **kwargs (extra keywords -> dict).
def deposit_all(*amounts):
    return sum(amounts)
print(deposit_all(100, 50, 25))                    # 175

def open_account(**fields):
    return {"is_active": True, "tags": [], **fields}
print(open_account(id=2, owner="Lin", balance=800.0))
```

## Section 4 — Exceptions

`raise` an error instead of returning a bad value or a silent `None` — a good error is **business logic made readable**: the type + message say which rule broke. The caller `try`s the risky code, `except`s the failure, runs `else` on success, and `finally` always (cleanup).

```python
# 4.1 · raise — a clear, typed error is business logic in plain words (not a bad value or silent None).
accounts = [
    {"id": 1, "owner": "Ada", "balance": 1500.0, "is_active": True},
    {"id": 3, "owner": "Sam", "balance": 40.0,   "is_active": True},
]

def withdraw(acct, amount):
    if amount > acct["balance"]:
        raise ValueError(f"insufficient funds: balance {acct['balance']}, asked {amount}")
    acct["balance"] -= amount
    return acct["balance"]

def find_account(accounts, acct_id):
    for a in accounts:
        if a["id"] == acct_id:
            return a
    raise LookupError(f"no account with id={acct_id}")

# Trigger the rule: Sam has 40, asks for 100.
try:
    withdraw({"id": 3, "owner": "Sam", "balance": 40.0}, 100)
except ValueError as err:
    print("declined:", err)
```

```python
# 4.2 · try / except / else / finally — handle the failure; else on success; finally always.
try:
    found = find_account(accounts, 1)
except LookupError as err:
    print("not found:", err)
else:
    print("found:", found["owner"])      # found: Ada
finally:
    print("lookup attempt done")

try:
    find_account(accounts, 99)
except LookupError as err:
    print("handled:", err)               # handled: no account with id=99
```

## Section 5 — The Zen of Python (clean code)

The Zen is about **clean, maintainable code** — read far more often than written. *"If you write code as cleverly as possible, you're not clever enough to debug it."* Here we refactor the same logic for clarity: honest **names**, explicit **type hints**, simple **flow** (truthiness, guard clauses).

```python
# 5.1 · Say what you mean — truthiness beats noisy comparisons.
tags = []
if len(tags) == 0: print("noisy: no tags")
if not tags:       print("clear: no tags")
```

```python
# 5.2 · Flatten the flow — a guard clause (early return) instead of deep nesting.
def describe_clever(a):                 # nested, hard to scan
    if a["is_active"]:
        if a["balance"] > 0:
            return "active, funded"
        else:
            return "active, empty"
    return "inactive"

def describe_clean(a):                  # flat: handle the edge first, then the main path
    if not a["is_active"]:
        return "inactive"
    return "active, funded" if a["balance"] > 0 else "active, empty"

acc = {"is_active": True, "balance": 1500.0}
print(describe_clever(acc), "|", describe_clean(acc))   # same answer, clearer code
```

```python
# 5.3 · Honest names + explicit type hints make a function self-documenting.
def calc(x): return sum(r["balance"] for r in x)                # what is x? what's returned?
def total_balance(accounts: list[dict]) -> float:              # name + hints tell the story
    return sum(a["balance"] for a in accounts)
print(total_balance([{"balance": 1500.0}, {"balance": 40.0}]))  # 1540.0
```

## Section 6 — Revision

All four moves on one little bank, written cleanly — the cumulative result of Sections 1–5. (The lab then builds the same shape on `Product`, not `BankAccount` — same patterns, different domain.)

```python
# 6.1 · The four moves on one little bank, written cleanly.
accounts: list[dict] = [
    {"id": 1, "owner": "Ada", "balance": 1500.0, "is_active": True,  "tags": ["primary"]},
    {"id": 2, "owner": "Lin", "balance": 0.0,    "is_active": False, "tags": []},
    {"id": 3, "owner": "Sam", "balance": 40.0,   "is_active": True,  "tags": []},
]

# functions + defaults + return + type hints (Sections 1, 3, 5)
def total_active_balance(accounts: list[dict]) -> float:
    return sum(a["balance"] for a in accounts if a["is_active"])

# control flow + truthiness (Section 2)
for a in accounts:
    status = "active" if a["is_active"] else "inactive"
    tag_text = ", ".join(a["tags"]) if a["tags"] else "no tags"
    print(a["owner"], status, "|", tag_text)

# exceptions: business logic as a readable error (Section 4)
def find_account(accounts: list[dict], acct_id: int) -> dict:
    for a in accounts:
        if a["id"] == acct_id:
            return a
    raise LookupError(f"no account with id={acct_id}")

try:
    print("total active:", total_active_balance(accounts))   # 1540.0
    print("found:", find_account(accounts, 2)["owner"])      # Lin
    find_account(accounts, 99)
except LookupError as err:
    print("handled:", err)                                   # handled: no account with id=99
```
